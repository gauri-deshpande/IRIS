```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.arange(1, 11)
y1 = np.random.randint(1, 10, size=10)
y2 = np.random.randint(1, 10, size=10)
plt.fill_between(x, y1, color="skyblue", alpha=0.4)
plt.fill_between(x, y2, color="salmon", alpha=0.5)
plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 25, 10, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'salmon', 'lightgreen', 'orange',
'yellow'])
plt.title('Pie Chart')
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=100)
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
x = np.random.randn(50)
y = 2 * x + np.random.randn(50)
plt.scatter(x, y, color='skyblue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
